```python
#Считываем из файла в список
lst = []
with open('file_extra.txt') as f:
    lst = f.read().splitlines()
```

```python
lst_amount = []
lst_gaps1 = []
lst_gaps2 = []
for i in range(0, len(lst), 3):
  lst_amount.append((int)(lst[i]))
  lst_gaps1.append((int)(lst[i+1]))
  lst_gaps2.append((int)(lst[i+2]))
```

```python
lst_gaps_per_scaffold1 = []
lst_gaps_per_scaffold2 = []
for i in range(0, len(lst_gaps1)):
  lst_gaps_per_scaffold1.append(1.0*lst_gaps1[i]/lst_amount[i])
  lst_gaps_per_scaffold2.append(1.0*lst_gaps2[i]/lst_amount[i])
```
